```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask') #한국어 모델

sentences = ["안녕하세요?", "한국어 문장 임베딩을 위한 버트 모델입니다."] #문장
embeddings = model.encode(sentences) #문장 encoding

print(embeddings) 
```

```python
df = pd.read_csv('data.csv')
#data 불러오기
#data.csv < 내용 
#-----------------------------------------------------------
#     	구분 ㅣ	유저 ㅣ	응답
#    0ㅣ감정ㅣ	힘들다ㅣ	무슨일이 힘들었어?
#    1ㅣ감정ㅣ	배고파ㅣ	그럼 먹고싶은게 뭐야?
#    2ㅣ감정ㅣ	눈물나ㅣ	울지마 누가 힘들게 했어 ㅜ
#    3ㅣ감정ㅣ	기쁘다ㅣ	너가 기뻐하니 나도 기뻐
#    4ㅣ감정ㅣ	고통스러워ㅣ	고통스럽구나 ㅜ 나도 고통스럽네

df.head()
```

```python
model.encode(df.loc[0, 'name'])
#유저 대화 encoding 
```

```python
df['embedding'] = pd.Series([[]] * len(df)) # dummy
#유저 대화를 encoding 해서 열추가
df['embedding'] = df['name'].map(lambda x: list(model.encode(x)))
#유저 응답 배열을 열에 추가해주기 
df.head()
```

```python
df.to_csv('data.csv', index=False)
#data.csv 덮어쓰기 저장
```

```python
text = '플라디아이 네일아트 21호' #유저 발화(유저가 쓴 채팅)

embedding = model.encode(text) 
```

```python
df['distance'] = df['embedding'].map(lambda x: cosine_similarity([embedding], [x]).squeeze())

df.head()
```

```python
answer = df.loc[df['distance'].idxmax()]

print('구분', answer['number'])
print('유사한 질문', answer['name'])
print('챗봇 답변', answer['sell'])
print('유사도', answer['distance'])
```
